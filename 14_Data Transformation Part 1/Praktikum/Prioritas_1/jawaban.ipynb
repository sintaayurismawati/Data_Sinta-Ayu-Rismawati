* Normalization (min to max)

```python
import pandas as pd

data = pd.read_csv('house_listings.csv')

data['price'] = data['price'].str.replace(' ','')
data['price'] = pd.to_numeric(data['price'], errors='coerce')
min_price = data['price'].min()
max_price = data['price'].max()

price_min_max_scaled_value = []

for price in data['price']:
    min_max = (price-min_price)/(max_price-min_price)
    price_min_max_scaled_value.append(min_max)

data['price_1m2'] = data['price_1m2'].str.replace('AZN/m²','')
data['price_1m2'] = data['price_1m2'].str.replace(' ','')
data['price_1m2'] = pd.to_numeric(data['price_1m2'], errors='coerce')
min_price_1m2 = data['price_1m2'].min()
max_price_1m2 = data['price_1m2'].max()

price_1m2_min_max_scaled_value = []

for price_per_m in data['price_1m2']:
    min_max = (price_per_m-min_price_1m2)/(max_price_1m2-min_price_1m2)
    price_1m2_min_max_scaled_value.append(min_max)

normalized_data = pd.DataFrame({'price': data['price'], 'price_min_max_scaled_value': price_min_max_scaled_value, 'price/m^2':data['price_1m2'], 'price_1m2_min_max_scaled_value':price_1m2_min_max_scaled_value })
normalized_data.head(20)
```

* Agregasi Data

```python
data = pd.read_csv('house_listings.csv')
data['price'] = data['price'].str.replace(' ','')
data['price'] = pd.to_numeric(data['price'], errors='coerce')

grouped_data = data.groupby(['room_number', 'category'])['price']

aggregated_data_mean = grouped_data.mean()
aggregated_data_median = grouped_data.median()
aggregated_data_modus = grouped_data.apply(lambda x: x.mode().iloc[0])

aggregated_data = pd.DataFrame({'Rata-rata': aggregated_data_mean, 'Median': aggregated_data_median,'Modus': aggregated_data_modus})
aggregated_data
```
